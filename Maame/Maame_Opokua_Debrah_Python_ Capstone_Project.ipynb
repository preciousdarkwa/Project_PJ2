# 1. 0 Introduction

This dataset contains data on used German Cars that have been listed on eBay.

## 1.1 Dataset Dictionary

* dateCrawled - When this ad was first crawled. All field-values are taken from this date.  
* name - Name of the car.
* seller - Whether the seller is private or a dealer.  
* offerType - The type of listing  
* price - The price on the ad to sell the car.  
* abtest - Whether the listing is included in an A/B test.  
* vehicleType - The vehicle Type.  
* yearOfRegistration - The year in which the car was first registered.  
* gearbox - The transmission type.  
* powerPS - The power of the car in PS.  
* model - The car model name.
* kilometer - How many kilometers the car has driven.  
* monthOfRegistration - The month in which the car was first registered.  
* fuelType - What type of fuel the car uses.  
* brand - The brand of car to sell  
* notRepairedDamage : if the car has a damage which is not repaired yet. Yes or no  
* dateCreated : the date for which the ad on Ebay was created  
* nrOfPictures : number of pictures in the ad  
* postalCode : code that shows the location of the car  
* lastSeenOnline : when the crawler saw this ad last online

### 6.3 Brands of Cars with the highest prices not considering the outliers

```python

Q1 = cars['Price'].quantile(0.25)
Q3 = cars['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

cars_no_outliers = cars[(cars['Price'] >= lower_bound) & (cars['Price'] <= upper_bound)]

top_brands = cars_no_outliers.groupby('Brand')['Price'].mean().sort_values(ascending=False).head(10).index

top_brands_data = cars_no_outliers[cars_no_outliers['Brand'].isin(top_brands)]

plt.figure(figsize=(16, 8))
barplot = sns.barplot(data=top_brands_data, x='Brand', y='Price', palette='viridis')

for p in barplot.patches:
    barplot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='right', va='center', xytext=(0, 15), textcoords='offset points', rotation='horizontal')

plt.title('Top 10 Car Brands with the Highest Prices')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
```

### 8. Correlation Map looking at the relations between Price, Kilometers, Horse Power and Car Age

```python
correlation_matrix = cars[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Correlation Values:

* The numerical values in each cell of the heatmap represent the correlation coefficient between two variables. The coefficient ranges from -1 to 1, where:  
* 1 indicates a perfect positive correlation (as one variable increases, the other also increases),  
* -1 indicates a perfect negative correlation (as one variable increases, the other decreases),  
* 0 indicates no correlation.

### Interpretation:

* A strong positive correlation suggests that as one variable increases, the other variable tends to increase as well.  
* A strong negative correlation suggests that as one variable increases, the other variable tends to decrease.  
* A correlation close to zero indicates a weak or no linear relationship.  
  
    

* **From the Heatmap**, we realize that as the age of the car increases (gets old), the price of the car decrease and vice versa.  
* The price of cars decreases when the kilometers the car has travelled is high  
* As the car's horse power increases its price increases meaning the older the car gets the better it performs. This relationship though is not strong so its subject to further analysis.

## 9. Conclusion

### Price Distribution:

The distribution of car prices in the dataset is right-skewed, indicating that the majority of cars have lower prices, with a few outliers having significantly higher prices.

### Vehicle Type Distribution:

The dataset consists of various vehicle types, with Limousine and Small Cars being the most common types. This suggests a diverse range of vehicles available for sale.  
### Year of Registration vs. Price:

There is a noticeable negative correlation between the year of registration and car prices. Generally, newer cars tend to have higher prices, while older cars are priced lower.  
### Brand vs. Price:

Prices vary significantly among different car brands.Brands such as Porsche and Sonstige Autos tend to have higher average prices compared to other brands.  
### Fuel Type Distribution:

The majority of cars in the dataset use Benzine as a fuel type. Diesel and lpg vehicles also have a notable presence, reflecting the diversity of fuel types in the used car market.  
### Relationship between Distance and Price:

There is a negative correlation between distance covered and car prices. Cars with lower distance coverage tend to be priced higher, indicating that distance driven is an important factor influencing the resale value of a car.  
### Gearbox Type Distribution:

The dataset contains a mix of manual and automatic gearboxes. Automatic cars, on average, have higher prices compared to manual cars though most of the cars in the dataset is manual.  
### Damaged vs. Non-damaged Cars:
A substantial proportion of cars in the dataset exhibits some degree of damage. On average, damaged cars are priced lower than their non-damaged counterparts, underscoring the considerable impact of vehicle condition on pricing.  
### Correlation Heatmap:

The correlation heatmap reveals that certain pairs of features, such as year of registration and kilometers, have a negative correlation. This reinforces the idea that newer cars with lower mileage tend to have higher prices.  
### Top Brands and Models:

BMW, Volkswagen, and Mercedes Benz are among the top brands with a high number of listings. Within these brands, specific models like the "Golf" and "3 ER" are particularly popular.


## Summary

When considering factors that influence used car prices, variables such as the year of registration, brand, horse power, distance covered(kilometers), and damage status play crucial roles.  
These insights can be valuable for both buyers and sellers in understanding market trends and making informed decisions.

## Recommendations

## Buyers

#### 1. Consider Your Budget:   
 Given the right-skewed distribution of prices, it's crucial to have a clear budget in mind. Understand the trade-offs between factors like age, distance_covered(kilometers), and brand to find the best value for your budget.

#### 2. Prioritize Vehicle Condition:   
The analysis showed that damaged cars tend to be priced lower. If you're open to cars with minor damage, you might find a good deal. However, always assess the extent of the damage and potential repair costs.

####  3. Explore Diverse Brands:   
While luxury brands like BMW and Mercedes-Benz the go to brand for a lot of buyers, don't overlook other reliable brands. Volkswagen, Opel, and Ford offer a range of models with varying price point and other factors.

## Sellers

### 1. Competitive Pricing:  
Understand the market trends revealed in the analysis. Price your car competitively based on factors like brand, model, year of registration, and mileage.

If your car has low mileage, is in excellent condition, or has unique features, make sure to highlight these in your listing. Buyers often look for specific attributes that add value.

Transparency is key. Provide accurate and detailed information about the car's history, maintenance, and any repairs. This builds trust with potential buyers.
