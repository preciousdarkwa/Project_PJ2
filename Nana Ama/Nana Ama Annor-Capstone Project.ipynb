```python
# converting data types

columns_to_change = ["InvoiceDate", "Customer ID"]

for column in columns_to_change:
    if column == "InvoiceDate":
        df1[column] = pd.to_datetime(df1[column])
    elif column == "Customer ID":
        df1[column] = df1[column].astype("object")
        
df1.dtypes
```

```python
# finding the number of unique countries mentioned to determine if it can be visualised
df1["Country"].nunique()
```

```python
# finding the total amount of revenue generated in each country
# creating a data frame of countries and total revenue and renaming columns appropriately
# Picking just the top ten

df1_total_revenue = df1.groupby("Country")["Price"].sum().to_frame()
df1_total_revenue = df1_total_revenue.rename(columns = {"Price" : "Total Revenue"})
df1_total_revenue = df1_total_revenue.sort_values(by="Total Revenue", ascending=False)
df1_total_revenue_top_10 = df1_total_revenue.head(10)
df1_total_revenue_top_10
```

```python
# plotting a horizontal bar chart of countries and their total revenue

plt.figure(figsize=(15, 15))
bars = plt.barh(df1_total_revenue_top_10.index, df1_total_revenue_top_10["Total Revenue"], color="pink", edgecolor="black")

plt.ylabel("Country")
plt.xlabel("Total Revenue")
plt.title("Bar Chart Showing the Total Revenue for Top 10 Countries")

# Add data labels on each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():,.2f}', 
             va='center', ha='left', fontsize=10, color='black')

plt.show()
```

```python
# now visualising total revenue made in each year

# Group by the year and calculate the total revenue for each year
df1_year = df1.groupby(df1['InvoiceDate'].dt.year)['Price'].sum().to_frame()

# Rename the columns
df1_year = df1_year.rename(columns={"InvoiceDate": "Year", "Price": "Total Revenue"})

# Sort the DataFrame by total revenue in descending order
df1_year = df1_year.sort_values(by="Total Revenue", ascending=False)

pd.DataFrame(df1_year)
```

```python
# Plotting a pie chart for the total revenue for each year

# Specify different shades of pink for the pie chart
custom_colors = ['#FFD1DC', '#FFB6C1', '#C71585']

# Plotting a pie chart with different shades of pink
plt.pie(df1_year['Total Revenue'], labels=df1_year.index, autopct='%1.1f%%', startangle=140, colors=custom_colors, wedgeprops=dict(edgecolor='black', linewidth=1.5))

# Equal aspect ratio ensures that the pie is drawn as a circle
plt.axis('equal')

# Set the title
plt.title('Distribution of Total Revenue Across Years')

# Show the pie chart
plt.show()
```

```python
# Revenue generated per month

df1.index = pd.to_datetime(df1.index)

# Create a new column for revenue
df1['Revenue'] = df1['Quantity'] * df1['Price']

# Resample by month and sum the revenue
monthly_revenue = df1.resample('M').sum()

plt.figure(figsize=(10, 6))
line_plot = plt.plot(monthly_revenue['Revenue'], marker='o', linestyle='-', color='#FF00FF', label='Monthly Revenue')

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

plt.title('Monthly Revenue Trends')
plt.xlabel('Date')
plt.ylabel('Revenue Generated')
plt.legend()
plt.show()
```

```python
# Sales Trends over Months
df1.index = pd.to_datetime(df1.index)
monthly_sales = df1.resample('M').sum()

plt.figure(figsize=(10, 6))
line_plot = plt.plot(monthly_sales['Quantity'], marker='o', linestyle='-', color='pink', label='Monthly Sales')

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.legend()
plt.show()
```

```python
# Group by product description and sum the quantities
product_quantities = df1.groupby('Description')['Quantity'].sum().reset_index()

# Convert the result to a DataFrame
df1_product_quantities = pd.DataFrame(product_quantities, columns=['Description', 'Quantity'])

# sorting on the quantities
df1_product_quantities_sorted = df1_product_quantities.sort_values(by='Quantity', ascending=False)

# fetching the top 10 best selling products
df1_product_quantities_top_10 = df1_product_quantities_sorted.head(10)

pd.DataFrame(df1_product_quantities_top_10)
```

```python
# Plotting a bar chart of the bestselling products
plt.figure(figsize=(10, 6))
plt.bar(df1_product_quantities_top_10['Description'], df1_product_quantities_top_10['Quantity'], color='hotpink', edgecolor='black')
plt.xlabel('Product Description')
plt.ylabel('Quantity Sold')
plt.title('Top 10 Best-Selling Products')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Ensure labels fit within the figure area
plt.show()
```

```python
# Customer Analysis - Distribution of Customers for Top 5 Countries with Data Labels and Gradient Color
top_countries = df1['Country'].value_counts().head(5).index

# Create a color map for the gradient (dark to light pink)
colors = sns.color_palette("RdPu", n_colors=5)

plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Country', data=df1[df1['Country'].isin(top_countries)], order=top_countries, palette=colors)

# Add data labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10, color='black')

plt.title('Distribution of Customers for Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')

# Set a gradient color legend
gradient_legend = [plt.Line2D([0], [0], color=colors[i], lw=4) for i in range(5)]
plt.legend(gradient_legend, top_countries, title='Countries', loc='upper right')

plt.show()
```
