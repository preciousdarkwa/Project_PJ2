```python
#Transaction Quantity Distribution
bin_edges = [i for i in range(1, 5)] 
sns.histplot(coffee_sales['transaction_qty'], bins=bin_edges)
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2., p.get_height(), f'{p.get_height():.0f}',
             ha='center', va='bottom')
plt.yticks([])
plt.show()
```

```python
# Bar Plot of Total Sales by Product Category
plt.figure(figsize=(9, 7))
sns.barplot(x='product_category', y='transaction_qty', data=coffee_sales, estimator=sum)
plt.title('Bar Plot of Total Sales by Product Category')
plt.xticks(rotation=45)
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2., p.get_height(), f'{p.get_height():.0f}',
             ha='center', va='bottom')
plt.yticks([])
plt.show()
```

```python
#total sales by store location
sns.barplot(x='store_location', y='transaction_qty', data=coffee_sales, estimator=sum)
plt.title('Bar Plot of Total Sales by Store Location')
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2., p.get_height(), f'{p.get_height():.0f}',
             ha='center', va='bottom')
plt.yticks([])
plt.show()
```

```python
#Correlation Matrix of Numerical Columns
numeric_columns = coffee_sales.select_dtypes(include=['float64', 'int64']).columns
filtered_columns = [col for col in numeric_columns if not col.endswith("id")]
correlation_matrix = coffee_sales[filtered_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar Plot of Product Distribution
sns.countplot(x='product_category', data=coffee_sales, order=coffee_sales['product_category'].value_counts().index)
plt.title('Bar Plot of Product Distribution')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2., p.get_height(), f'{p.get_height():.0f}',
             ha='center', va='bottom')
plt.yticks([])
plt.show()
```

```python
#count of transactions by store location
sns.countplot(x='store_location', data=coffee_sales)
plt.title('Count Plot of Transactions by Store Location')
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2., p.get_height(), f'{p.get_height():.0f}',
             ha='center', va='bottom')
plt.yticks([])
plt.show()
```

```python
#Calculate total sales for each product type
product_type_sales = coffee_sales.groupby('product_type')['transaction_qty'].sum().reset_index()

#Identify the top 10 selling product types
top_10_product_types = product_type_sales.nlargest(10, 'transaction_qty')

#Create a bar plot to visualize the top 10 selling product types
plt.figure(figsize=(13, 7))
sns.barplot(x='transaction_qty', y='product_type', data=top_10_product_types, palette='viridis')
plt.title('Top 10 Best Selling Product Types')
plt.xlabel('Total Sales')
plt.ylabel('Product Type')
for p in plt.gca().patches:
    plt.text(p.get_width(), p.get_y() + p.get_height() / 2., f'{p.get_width():,.0f}',
             ha='left', va='center')
plt.xticks([])
plt.show()
```

```python
#Calculate total sales for each product type
product_type_sales = coffee_sales.groupby('product_type')['transaction_qty'].sum().reset_index()

# Identify the bottom 10 selling product types
bottom_10_product_types = product_type_sales.nsmallest(10, 'transaction_qty')

# Reverse the order of the dataframe for descending order
bottom_10_product_types = bottom_10_product_types[::-1]

# Create a bar plot to visualize the bottom 10 selling product types
plt.figure(figsize=(13, 7))
sns.barplot(x='transaction_qty', y='product_type', data=bottom_10_product_types, palette='viridis')
plt.title('Bottom 10 Selling Product Types')
plt.xlabel('Total Sales')
plt.ylabel('Product Type')
for p in plt.gca().patches:
    plt.text(p.get_width(), p.get_y() + p.get_height() / 2., f'{p.get_width():,.0f}',
             ha='left', va='center')
plt.xticks([])
plt.show()

```

```python
# Aggregate by time and calculate total transaction quantity
time_based_sales = coffee_sales.groupby('transaction_time')['transaction_qty'].sum().reset_index()

# Identify the top 10 time periods
top_10_time_periods = time_based_sales.nlargest(10, 'transaction_qty')

# Create a line plot to visualize the top 10 time periods
plt.figure(figsize=(12, 6))
sns.lineplot(x='transaction_time', y='transaction_qty', data=top_10_time_periods, marker='o')
plt.title('Top 10 Time Periods Based on Transaction Quantity')
plt.xlabel('Transaction Time')
plt.ylabel('Total Transaction Quantity')
plt.show()
```

```python
#Store Location Clustering based on Sales
from sklearn.cluster import KMeans

store_location_data = coffee_sales.groupby('store_location')['transaction_qty'].sum().reset_index()

# Perform K-means clustering
X = store_location_data[['transaction_qty']]
kmeans = KMeans(n_clusters=3) 
store_location_data['cluster'] = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=store_location_data, x='store_location', y='transaction_qty', hue='cluster', palette='viridis')
plt.title('Store Location Clustering based on Sales')
plt.xlabel('Store Location')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Time series analysis
from statsmodels.tsa.arima.model import ARIMA

time_series_data = coffee_sales.groupby('transaction_date')['transaction_qty'].sum()

# Fit ARIMA model
model = ARIMA(time_series_data, order=(1, 1, 1))
model_fit = model.fit()

# Make predictions for future sales
forecast_steps = 12  # Adjust as needed
forecast = model_fit.get_forecast(steps=forecast_steps)

# Print or visualize the forecast results
print(forecast.predicted_mean)
```

# Conclusion

The analysis of the "coffee_sales" dataset has provided valuable insights into transaction patterns, product sales, and location-based trends. The key findings are summarized as follows:

### Transaction Quantity Distribution:

The histogram of transaction quantities revealed that the majority of transactions (58.5%) consisted of a quantity of 1, followed by 39.3% for a quantity of 2. Transactions with quantities 3, 4, 6, and 8 occurred less frequently, with percentages ranging from 0.002% to 2.1%.

### Product Category Sales:

Utilizing a bar plot, it was observed that "Coffee" emerged as the highest-selling product category, with a total of 89,250 transactions, followed by "Tea" at 69,737 and "Bakery" at 23,214. On the other hand, "Packaged Chocolate" and "Branded" were the lowest-selling categories, with sales of 487 and 776, respectively.

### Store Location Clustering based on Sales:

Following a k-cluster analysis on the dataset, the examination of total sales by location identified "Lower Manhattan" with the highest sales at $71,742, followed  closely  by "Hell's Kitchen" at  $71,737 and "Astoria" at $70,991. 

### Product Category Distribution:

Using a bar plot to visualize the distribution of product categories, it was found that "Coffee" appeared most frequently in transactions (58,416 times), followed by "Tea" (45,449) and "Bakery" (22,796). Conversely, "Packaged Chocolate" and "Branded" had the lowest occurrences at 487 and 747, respectively.

### Transaction Counts by Location:

"Hell's Kitchen" recorded the highest number of transactions (50,735), followed by "Astoria" (50,599) and "Lower Manhattan" (47,782).

### Best and Worst Selling Product Types:

The top-selling product types were "Brewed Chai Tea" and "Gourmet Brewed Coffee," with transaction quantities of 26,250 and 25,973, respectively. Conversely, the worst-selling product types were "Green Beans" (134) and "Green Tea" (159).

### Optimal Transaction Time:

Generally, optimal transaction time was in the morning.

### Overall Sales:

The cumulative total sales for the dataset amounted to 214,470.
